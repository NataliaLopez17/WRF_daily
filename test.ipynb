```python
import netCDF4 as nc
import pandas as pd
import xarray as xr
import datetime
import glob
import os
import datetime
```

```python
#This line reads all the files in the directory and puts them in a list
netcdf_list = glob.glob('NETCDF Files/*.nc', recursive=True)

#This line initializes empty lists to store the data according to the years and models
datasets_CNRM_1985 = []
datasets_CNRM_2040 = []
datasets_CESM_1985 = []
datasets_CESM_2040 = []

#This line reads the data from the netcdf list and puts each file in a list according to the year and model
for i in netcdf_list:
    # print(netcdf_ds)
    if i.find('CNRM') != -1 and i.find('1985') != -1:
        datasets_CNRM_1985.append(i)
    elif i.find('CNRM') != -1 and i.find('2040') != -1:
        datasets_CNRM_2040.append(i)
    elif i.find('CESM') != -1 and i.find('1985') != -1:
        datasets_CESM_1985.append(i)
    else:
        datasets_CESM_2040.append(i)
print(datasets_CESM_2040)
```

```python
#dates to use in the loop
first_date = datetime.datetime(1985, 1, 1)
end_date = datetime.datetime(2005, 12, 31)
day = datetime.timedelta(days=1)
days = pd.date_range(start=first_date, end=end_date, freq='D')

#path to store the files
path = 'D:/Users/viaje/OneDrive - University of Puerto Rico/Harmsen_Work_Jan_May_2022/WRF_daily/NETCDF Files csv/CNRM-1985-2005/'

#loops over the list containing the different files from the CNRM model with the 1985 data
for i in datasets_CNRM_1985:
    for d in days:
        #splits the name of the file to get the name of the parameter
        my_list = i.split('\\')[-1]
        name = my_list.split('_')[0]
        
        #opens the file as a dataset
        netcdf_ds = xr.open_dataset(i)
        #selects the data to be used using a time range since the data starts on 2039 and not 2040.
        netcdf_ds = netcdf_ds.sel(Time=slice(first_date, end_date))
        netcdf_df = netcdf_ds.to_dataframe()
        netcdf_df = netcdf_df.reset_index()
        netcdf_df = netcdf_df.drop(columns=['bnds', 'south_north', 'west_east', 'Time_bnds'])
        netcdf_df = netcdf_df.loc[netcdf_df['Time'] == d]
        dt_object = d.to_pydatetime()
        dt_object = dt_object.strftime('%Y-%m-%d')
        netcdf_df.to_csv(os.path.join(source_folder, name,
                        '{dt_object}.csv'.format(dt_object=dt_object)))
```

```python
first_date = datetime.datetime(2040, 1, 1)
end_date = datetime.datetime(2060, 12, 31)
day = datetime.timedelta(days=1)
days = pd.date_range(start=first_date, end=end_date, freq='D')

CESM_2040_csvs = glob.glob('NETCDF Files csv/CESM-2040-2060/*.csv', recursive=True)
source_folder = 'NETCDF Files csv/CESM-2040-2060/CESM-2040-2060-daily/'

for i in CESM_2040_csvs:
    for d in days:
        file_name = i.split('\\')[-1]
        parameter = file_name.split('.')[0]
        df = pd.read_csv(i)
        df = df.drop(columns=['Unnamed: 0'])
        df = df.reset_index()
        df = df.drop(columns=['index'])
        df['Time'] = df['Time'].astype('datetime64[ns]')
        df = df.loc[df['Time'] == d]
        # dt_object = dt_object.strftime('%Y-%m-%d')
        df.to_csv(os.path.join(source_folder, parameter,'f{d}.csv'))
```
